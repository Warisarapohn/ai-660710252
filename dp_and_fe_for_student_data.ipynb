#-Data preparation

1.import library ที่จำเป็น

```python
import pandas as pd
import numpy as np
import math
from datetime import datetime, timedelta
```

2.อ่านไฟล์

```python
df = pd.read_csv('student.csv')
```

3.ดูข้อมูลในตาราง

```python
df.head(20)
```

4.ทำสำเนา

```python
df_copy = df.copy()
```

5.ตรวจดูแต่ละหลักว่ามีค่าว่างเท่าไรบ้าง

```python
df_copy.isnull().sum()
```

6.ทำการลบแถวที่ชื่อเป็นค่าว่าง

```python
df_copy.dropna(subset=['ชื่อ'], inplace=True)
df_copy
```

7หาค่าเฉลี่ยอายุนักเรียนทั้งหมดในตารางเอามาแทนในค่าว่าง และวนloopว่ามีค่าในหลักไหนของ 'ชื่อ' ที่มากกว่า 30 แล้วแทนด้วยค่าเฉลี่ย

```python
mean = round(df_copy['อายุ'].mean())
df_copy['อายุ'].fillna(mean, inplace=True)
for i, age in df_copy['อายุ'].items():
    if age  > 30:
        df_copy.at[i, 'อายุ'] = mean

df_copy
```

8.ทำการจับคู่คำที่ความหมายเดียวกันให้ค่าเท่ากับค่า ค่าหนึ่งไปเลยเพื่อให้ข้อมูลมีความสม่ำเสมอ

```python
gender_mapping = {
    'ชาย': 'ชาย',
    'หญิง': 'หญิง',
    'F': 'หญิง',
    'M': 'ชาย',
    'ญ': 'หญิง',
    'ช': 'ชาย',
    'Female': 'หญิง',
    'Male' : 'ชาย'
}
```

9.แล้วนำค่าที่เราจับคู่มาปรับใช้กับตาราง แล้วค่าว่างให้เราแทนด้วย ไม่ระบุ

```python
df_copy['เพศ']=df_copy['เพศ'].map(gender_mapping)
df_copy['เพศ'].fillna('ไม่ระบุ', inplace=True)

df_copy
```

10.จับคู่คำที่มีความหมายเดียวกัน ให้ค่าเท่ากับค่าๆหนึ่ง แล้วให้ได้ข้อมูลที่สม่ำเสมอ

```python
faculty_mapping = {
    'วิทย์': 'วิทยาศาสตร์',
    'คณะวิทยาศาสตร์' : 'วิทยาศาสตร์',
    'Science': 'วิทยาศาสตร์',
    'Sci' : 'วิทยาศาสตร์',
    'วิทยาศาสตร์': 'วิทยาศาสตร์',
    'ศิลปศาสตร์': 'ศิลปศาสตร์',
    'Arts': 'ศิลปศาสตร์',
    'Liberal Arts': 'ศิลปศาสตร์',
    'ศิลป์' : 'ศิลปศาสตร์',
    'อุตสาหกรรม': 'วิศวกรรมศาสตร์',
    'วิศวกรรมศาสตร์': 'วิศวกรรมศาสตร์',
    'วิศวกรรม': 'วิศวกรรมศาสตร์',
    'วิศวกรรมอุตสาหกรรม': 'วิศวกรรมศาสตร์',
    'วิศวะ': 'วิศวกรรมศาสตร์',
    'วิศวฯ': 'วิศวกรรมศาสตร์',
    'Engineering': 'วิศวกรรมศาสตร์'
}
```

11.เอาค่าที่เราจับคู่มาปรับใช้กับตารารง แล้วค่าว่างให้แทนด้วย ไม่ระบุ

```python
df_copy['คณะ'] = df_copy['คณะ'].map(faculty_mapping)
df_copy['คณะ'].fillna('ไม่ระบุ', inplace=True)

df_copy
```

12.หาค่าเฉลี่ยเกรดนักเรียนทั้งหมดในตารางแล้วเอามาแทนในค่าว่าง วนloopว่ามีค่าในหลักไหนของ 'GPA' ที่มากกว่า 4.00 และมีค่าน้อยกว่า 0.00 แล้วแทนด้วยค่าเฉลี่ย

```python
GPAmean = round(df_copy['GPA'].mean())
df_copy['GPA'].fillna(GPAmean, inplace=True)
for i, GPA in df_copy['GPA'].items():
    if GPA  > 4.00:
        df_copy.at[i, 'GPA'] = GPAmean

    if GPA < 0.00:
        df_copy.at[i, 'GPA'] = GPAmean

df_copy
```

13.หาค่าเฉลี่ยของน้ำหนัก, ค่าเฉลี่ยของส่วนสูง และ ชั่วโมงเรียนแล้วแทนในค่าว่าง

```python
heighmean = round(df_copy['ส่วนสูง_cm'].mean())
df_copy['ส่วนสูง_cm'].fillna(heighmean, inplace=True)

Weighmean = round(df_copy['น้ำหนัก_kg'].mean())
df_copy['น้ำหนัก_kg'].fillna(Weighmean, inplace=True)

hourmean = round(df_copy['ชั่วโมงเรียน_ต่อสัปดาห์'].mean())
df_copy['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(hourmean, inplace=True)

df_copy
```

3.สร้างตารางแยกจากของเก่าเพื่อให้แสดงผลจะได้ไม่สับสน

```python
student_health = pd.DataFrame({
    'น้ำหนัก_kg': df_copy['น้ำหนัก_kg'],
    'ส่วนสูง_cm': df_copy['ส่วนสูง_cm']
})
```

4.สร้าง feature ให้ Ai ช่วยคิดคำนวน BMI เพื่อนำมาจัดกลุ่ม

```python
student_health['BMI'] = student_health['น้ำหนัก_kg'] / (student_health['ส่วนสูง_cm']/100)**2
student_health['สุขภาพ'] = pd.cut(student_health['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ต่ำกว่าเกณฑ์', 'สมส่วน', 'เกินมาตรฐาน', 'เป็นโรคอ้วน'])

student_health
```

5.สร้างตารางแยกจากของเก่าเพื่อให้แสดงผลจะได้ไม่สับสน



```python
student_grade = pd.DataFrame(df_copy['GPA'])
```

6.สร้าง feature ให้ Ai ช่วยจัดกลุ่มเกรดเฉลี่ย

```python
student_grade['เกรด'] = pd.cut(student_grade['GPA'],
                                   bins=[0, 0.99, 1.29 ,1.59, 2.29, 2.59, 3.29, 3.59, 4.00],
                                   labels=['F','D','D+', 'C', 'C+', 'B','B+', 'A'])
student_grade
```

7.สร้างตารางแยกจากของเก่าเพื่อให้แสดงผลจะได้ไม่สับสน

```python
student_date = pd.DataFrame(df_copy['วันที่ลงทะเบียน'])
```

8.สร้าง Feature จากวันเวลา ให้ Ai ช่วยดูว่าวันที่นักเรียนลงทะเบียนคือวันอะไรในสัปดาห์

```python
student_date['วันที่ลงทะเบียน'] = pd.to_datetime(student_date['วันที่ลงทะเบียน'])
student_date['วันในสัปดาห์'] = student_date['วันที่ลงทะเบียน'].dt.day_name()
student_date
```
